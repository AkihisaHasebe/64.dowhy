## 8. まとめと使い分けガイドライン

### SHAP と因果探索の本質的な違い

| 観点 | SHAP (XAI) | 因果探索 (LiNGAM等) |
|:---|:---|:---|
| **問い** | 「モデルの予測に何が寄与したか？」 | 「何が何を引き起こすか？」 |
| **対象** | 学習済みモデル（ブラックボックス） | データの生成過程 (DGP) |
| **擬似相関の扱い** | 区別できない（予測に有用なら高評価） | 区別できる（直接効果のみ検出） |
| **前提条件** | 予測モデルが必要 | 線形性・非ガウス性等の仮定（手法による） |
| **非線形対応** | 対応可（モデル依存） | LiNGAM: 不可 / PC+KCI: 可 |
| **適用場面** | モデルの説明・デバッグ | 介入計画・政策決定 |

### 使い分けの指針

**SHAP を使うべき場面:**
- モデルがなぜその予測をしたかを説明したい
- モデルの挙動をデバッグしたい
- 規制対応のための説明可能性が必要
- 予測精度に貢献する特徴量を知りたい

**因果探索を使うべき場面:**
- 介入（操作）の効果を知りたい
- 「何を変えれば結果が変わるか」を知りたい
- 擬似相関を排除して真の要因を特定したい
- ビジネスアクションに繋げたい

### 危険な誤解

```
❌ 「SHAP重要度が高い = 原因である」
✅ 「SHAP重要度が高い = モデルの予測に寄与している（原因とは限らない）」

❌ 「SHAP重要度が高い特徴量を操作すれば結果が変わる」
✅ 「因果効果がある特徴量を操作すれば結果が変わる」

❌ 「LiNGAM で因果効果ゼロ = 因果関係なし」（非線形の場合は注意！）
✅ 「非線形関係の可能性がある場合は、カーネルCI検定ベースの手法も併用する」
```

### 推奨ワークフロー

```
1. 因果探索で変数間の因果構造を推定
   - 線形性が仮定できる → LiNGAM
   - 非線形の可能性がある → PC + KCI テスト
2. 因果構造に基づいて真の原因変数を特定
3. 真の原因変数のみで予測モデルを構築
4. SHAP でモデルの予測を説明
→ 因果的に正しく、かつ解釈可能なモデルが得られる
```
